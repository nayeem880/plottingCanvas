```python
# plot noisy data and their polynomial fit
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(15)
x = np.linspace(0,1,20)
y = np.cos(x)+ 0.3*np.random.rand(20)

p = np.poly1d(np.polyfit(x,y,3))
t = np.linspace(0,1,200)
plt.plot(x,y,'o',t,p(t),'--')
plt.show()
```

```python
# Plot noisy data and their polynomial fit in a Chebyshev basis

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(10)
x = np.linspace(-1,1,2000)
y = np.cos(x)+ 0.3*np.random.rand(2000)
p = np.polynomial.Chebyshev.fit(x,y,90)
t = np.linspace(-1,1,200)
plt.plot(x,y,"c.")
plt.plot(t,p(t),"r-",lw=3)
plt.show()
```

```python
# compute the mandelbrot fractal and plot it

import numpy as np
import matplotlib.pyplot as plt
from numpy import newaxis

def compute_mandelbrot(N_max, some_threshold, nx, ny):
    # A grid of c-values
    x = np.linspace(-2, 1, nx)
    y = np.linspace(-1.5, 1.5, ny)

    c = x[:,newaxis] + 1j*y[newaxis,:]

    # Mandelbrot iteration

    z = c
    for j in range(N_max):
        z = z**2 + c

    mandelbrot_set = (abs(z) < some_threshold)

    return mandelbrot_set

mandelbrot_set = compute_mandelbrot(50, 50., 601, 401)

plt.imshow(mandelbrot_set.T, extent=[-2, 1, -1.5, 1.5])
plt.gray()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(-np.pi,np.pi,256,endpoint=False)
x,c = np.cos(a), np.sin(a)

plt.plot(a,x,"r--")
plt.plot(a,c,"c.")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20),dpi=100)
plt.subplot(2,1,1)
a= np.linspace(-np.pi, np.pi, 256, endpoint=True)
x,c = np.sin(a),np.cos(a)
plt.plot(a,x,linestyle="--",linewidth=2,color="red")
plt.plot(a,c,linestyle="solid",linewidth=2,color="cyan")
#  supported values are '-', '--', '-.', ':', 'None', 
# ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.xlim(-4,4)
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

plt.ylim(-2,2)
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

plt.savefig("plotstat1.png",dpi=100)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc("text", usetex = False)
a= np.outer(np.arange(0,1,0.01),np.ones(10))
plt.figure(figsize=(15,10))
plt.subplots_adjust(top=0.8,bottom=0.05, left =0.01,right=0.99)
maps = [m for m in plt.cm.datad if not m.endswith("_r")]
maps.sort()
l= len(maps)+2
for i , m in enumerate(maps):
    plt.subplot(1,l,i+1)
    plt.axis("off")
    plt.imshow(a,aspect="auto",cmap=plt.get_cmap(m), origin="lower")
    plt.title(m,rotation=90,fontsize = 10,va ="bottom")
plt.show()
```
